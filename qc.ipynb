```python
import csv
import re
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly.plotly as py
import cufflinks as cf
 
```

```python
def read_inserts(filename):
    myfile = open(filename)
    sizes = []
    counts = [] 

    for  row in myfile:

         line = row.split()
         sizes.append(line[0])
         counts.append(int(line[1]))

 
    return (sizes, counts)
    
```

```python
sizes, counts = read_inserts('bgi_hist.txt')
%matplotlib inline
x = sizes
#plt.hist(x, normed=True, bins=30)
plt.plot(x, counts, '--')
plt.ylabel('Counts');
plt.xlabel('Insert Sizes')
```
