```python
def writelist (mylist, myfile): 
    fp =  open(myfile,'a+')
    for idx,item in enumerate(mylist):
            fp.write(item)
            fp.write('\t')
    fp.write('\n')
    fp.close()
```

```python
def common(list1, list2, list3, list4):
    common_list = [] 
    for i in list1: 
        if i in  (list2 and list3 and list4): 
                common_list.append(i)      
    return common_list
```

```python
def difference(list1, list2):
    unique_list = []
    for i in list1: 
        if i not in list2: 
                unique_list.append(i)
    return unique_list
```

```python
def exist (list1, list2): 
    found_list = []
    for i in list1: 
            if i not in found_list: 
                    found_list.append(i)
    for i in list2: 
            if i not in found_list: 
                    found_list.append(i)
    return found_list 
```

# Read annovar annotation file

```python
 def load_annovar(filename):
    rows = []
    exacall = []
    gnomall = []
    rs = []
    gene = []
    pathogenics = []
    functions = []
    exonicfunctions =[]
    muttype = []
    fp = open(filename, 'r')
    r = csv.reader(fp, delimiter='\t')
    for n, row in enumerate(r):
        #rs ,pos,_,ensg,_, enst,_,_,_,_,_,_,_,impact =row[:14]
        if n == 0: 
                #print (row[8])
                #print(row[16])
                #print(row[124])
                #print(row[24])
                #print(row[6])
                #print (row[5])
                #print(row[59])
            
                continue
        exonicfunctions.append(row[8])
        muttype.append(row[13])
        exacall.append(row[16])
        gnomall.append(row[124])
        rs.append(row[24])
        rows.append(row)
        gene.append(row[6])
        functions.append(row[5])
        pathogenics.append(row[59])
    return rows, functions, exonicfunctions, rs, gene,  exacall , gnomall, pathogenics,muttype
```
